```python
import matplotlib.pyplot as plt
import numpy as np
 
city=['SVM','Bayes','maxEnt','CNN','MMCNN']
Gender=['positive sample','negative sample']
pos = np.arange(len(city))
bar_width = 0.35
Happiness_Index_Male=[90,90,87,91,93]
Happiness_Index_Female=[87,82,88,87,89]
#axes = plt.gca()
#axes.set_ylim([80,100])
plt.axis([-0.5, 5,80, 100])
#plt.figure(figsize=(10,7))
plt.bar(pos,Happiness_Index_Male,bar_width,color='violet',edgecolor='black')
plt.bar(pos+bar_width,Happiness_Index_Female,bar_width,color='yellow',edgecolor='black',hatch='////')
plt.xticks(pos, city)
plt.ylabel('predict accurate %', fontsize=12)
plt.title('Accuracy of different model on dataset',fontsize=18)
plt.legend(Gender,loc=2)
plt.show()

```
